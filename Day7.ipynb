Create a simple line plot using Matplotlib to display the trend of a list of values over time.

```python
import matplotlib.pyplot as plt

values = [4, 8, 12, 16, 20, 24, 28]
time = range(len(values)) 

plt.plot(time, values, marker='o')

plt.title('Trend of Values Over Time')
plt.xlabel('Time')
plt.ylabel('Values')

plt.show()
```

 Plot a bar chart using Matplotlib to show the frequency of different categories in a dataset.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D','E']
frequencies = [12, 5, 25, 21, 28]

plt.bar(categories, frequencies, color='blue')

plt.title('Frequency of Different Categories')
plt.xlabel('Categories')
plt.ylabel('Frequency')

plt.show()
```

Create a box plot using Seaborn to show the distribution of values for different categories in a dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')

sns.boxplot(x='day', y='total_bill', data=data)

plt.title('Distribution of Total Bill Amounts by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')

plt.show()
```

Plot a heatmap using Seaborn to visualize the correlation matrix of a dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('iris')

numeric_data = data.select_dtypes(include=[float, int])

correlation_matrix = numeric_data.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Matrix Heatmap')

plt.show()
```

Use Matplotlib to create a subplot grid that displays multiple charts in a single figure.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = x ** 2

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y1, 'b-')
axs[0, 0].set_title('Sine Wave')

axs[0, 1].plot(x, y2, 'r-')
axs[0, 1].set_title('Cosine Wave')

axs[1, 0].plot(x, y3, 'g-')
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].set_ylim([-10, 10])  # Limiting y-axis for better visualization

axs[1, 1].plot(x, y4, 'm-')
axs[1, 1].set_title('Quadratic')

fig.suptitle('Multiple Subplots Example', fontsize=16)

fig.tight_layout(pad=3.0)

plt.show()
```

Customize the appearance of a Seaborn plot by changing the color palette, adding titles, and modifying axis labels.

```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore", message="use_inf_as_na option is deprecated and will be removed in a future version")
pd.set_option('mode.use_inf_as_na', True)

iris = sns.load_dataset('iris')

sns.set_palette('husl')

sns.pairplot(iris, hue='species', height=2.5)

plt.suptitle('Pairplot of Iris Dataset by Species', fontsize=16, fontweight='bold')

plt.show()
```

Create a violin plot using Seaborn to visualize the distribution of a dataset across different categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

sns.set_palette('muted')

plt.figure(figsize=(10, 6))
ax = sns.violinplot(x='species', y='sepal_length', data=iris)

ax.set_title('Distribution of Sepal Length by Species', fontsize=16, fontweight='bold')
ax.set_xlabel('Species', fontsize=14, fontweight='bold')
ax.set_ylabel('Sepal Length (cm)', fontsize=14, fontweight='bold')

plt.show()
```

```python
Combine Matplotlib and Seaborn to create a complex visualization, such as overlaying a KDE plot on a histogram.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

plt.figure(figsize=(10, 6))

sns.histplot(iris['sepal_length'], bins=20, kde=False, color='blue', label='Histogram')

sns.kdeplot(iris['sepal_length'], color='red', linewidth=2, label='KDE')

plt.title('Histogram and KDE of Sepal Length', fontsize=16, fontweight='bold')
plt.xlabel('Sepal Length (cm)', fontsize=14, fontweight='bold')
plt.ylabel('Density', fontsize=14, fontweight='bold')

plt.legend()

plt.show()
```
